```python
# Import the ast module to safely evaluate the string as a Python literal
import ast  

tours['description'] = tours['tour_data'].apply(lambda x: ast.literal_eval(x)['overview']['route_description'])

tours.head()
```

```python
tours
```

```python
tours['name'] = tours['name'].str.split(n=1).str[1]
tours
```

```python
adventures.columns
```

```python
adventures.drop(['_id'], axis=1, inplace=True)
```

```python
adventures.columns
```

```python
adventures = adventures.rename(columns={
    'adventureName': 'name',
    'coverImage': 'imageCover',
    'adventureDescription': 'description'
})

adventures.columns
```

```python
adventures.head()
```

```python
adventures['price'] = adventures['price'] / 100
adventures.head()
```

```python
tours.columns
```

```python
tours.drop(['ratingsAverage', 'ratingsQuantity', 'tour_data', 'getting_there_data'], axis=1, inplace=True)
```

```python
tours.columns
```

```python
combined = pd.concat([tours, adventures], ignore_index=True)

combined
```

```python
combined.to_csv('final_tours_and_adventures.csv', index=False)
```
