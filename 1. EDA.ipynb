```python
# Assuming category_ratings_final_shuffled is your shuffled DataFrame
category_ratings_final.to_csv('user_category_ratings_google_final.csv', index=False)
```

```python
# Shuffle the dataset
shuffled_category_ratings = category_ratings_final.sample(frac=1).reset_index(drop=True)

# Drop the 'user' column
shuffled_category_ratings_no_user = shuffled_category_ratings.drop('user', axis=1)

# Select any 139 rows
subset_shuffled_category_ratings = shuffled_category_ratings_no_user.head(139)

# Concatenate along axis 1
resulting_data = pd.concat([users_profile_data, subset_shuffled_category_ratings], axis=1)

# Print or use resulting_data as needed
resulting_data
```
