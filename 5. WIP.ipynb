```python
# Drop unnecessary columns
columns_to_drop = ['coverImage', 'categories', 'personalities', 'createdAt']
data = data.drop(columns=columns_to_drop)

# Combine 'accomodation_preference' and 'transport_preference' into a single column 'preference'
preference_columns = [col for col in data.columns if 'preference' in col]
data['preference'] = data[preference_columns].idxmax(axis=1)
data.drop(columns=preference_columns, inplace=True)
```

```python
data.head()
```

```python
data.columns
```

```python
# Drop 'preference' column
data = data.drop(columns=['preference'])
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Create 'rating' column using PCA
category_columns = ['safari', 'art', 'food', 'history', 'nature', 'outdoors', 'sports', 'culture', 'drinks', 'entertainment', 'wellness']
category_data = data[category_columns]

# Standardize the data
scaler = StandardScaler()
category_data_scaled = scaler.fit_transform(category_data)

# Apply PCA
pca = PCA(n_components=1)
data['rating'] = pca.fit_transform(category_data_scaled)

# Drop the original category columns
data = data.drop(columns=category_columns)
```

```python
# Remove commas and spaces, then convert 'avg_accomodation_cost' and 'avg_transport_cost' to int
data['avg_accomodation_cost'] = data['avg_accomodation_cost'].str.replace(',', '').str.replace(' ', '')
data['avg_transport_cost'] = data['avg_transport_cost'].str.replace(',', '').str.replace(' ', '')

# Convert to int
data['avg_accomodation_cost'] = pd.to_numeric(data['avg_accomodation_cost'], errors='coerce')
data['avg_transport_cost'] = pd.to_numeric(data['avg_transport_cost'], errors='coerce')

# Fill NaN values with 0
data['avg_accomodation_cost'] = data['avg_accomodation_cost'].fillna(0).astype(int)
data['avg_transport_cost'] = data['avg_transport_cost'].fillna(0).astype(int)
```
